```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi=np.pi
ts=np.arange(0,10,0.1)
dt=0.02
g=9.8
l=9.8
omiga=0.0
t=0.0
n=200000
theta=0
q=0.5
omiga_list=[]
theta_list=[]
time_list=[]
driven_force=1.2


for i in range(n):
    t=t+dt
    omiga=omiga-g/l*np.sin(theta)*dt-q*omiga*dt+driven_force*np.sin(2/3*t)*dt
    theta=theta+omiga*dt
    if np.abs(theta)>pi and theta>0:
        theta=theta-2*pi
    if np.abs(theta)>pi and theta<0:
        theta=theta+2*pi
    omiga_list.append(omiga)
    theta_list.append(theta)
    time_list.append(t)
    
    
    
plt.plot(time_list,theta_list)
plt.xlabel('underdamped with driven force')
```

```python
pi=np.pi
ts=np.arange(0,10,0.1)
dt=0.02
g=9.8
l=9.8
omiga=0.0
t=0.0
n=5000
nn=200
theta=0.2
q=0.5
omiga_list=[]
theta_list=[]
time_list=[]
ovt_list=[]
driven_force=1.4
sp_time_list=[]
sp_omiga_list=[]
sp_theta_list=[]


for i in range(n):
    t=t+dt
    omiga=omiga-g/l*np.sin(theta)*dt-q*omiga*dt+driven_force*np.sin(2/3*t)*dt
    theta=theta+omiga*dt
    if np.abs(theta)>pi and theta>0:
        theta=theta-2*pi
    if np.abs(theta)>pi and theta<0:
        theta=theta+2*pi
    ovt=omiga/theta
    omiga_list.append(omiga)
    theta_list.append(theta)
    time_list.append(t)
    ovt_list.append(ovt)
    
    for j in range(nn):
     if np.abs(t)-3*j*pi<0.02 and np.abs(t)-3*j*pi>0:
        sp_time_list.append(t)
        sp_omiga_list.append(omiga)
        sp_theta_list.append(theta)
    
    
plt.scatter(sp_theta_list,sp_omiga_list)
plt.xlabel('driven-force=1.4')
```
